```python
from PIL import Image
crop = Image.open("mountains.jpg")
img = Image.new("RGBA", (crop.width*5, crop.height*3), "red")
copy = crop.copy() # kopietje maken om te kunnen plakken

for row in range(0,3):
    for col in range(0,5):
        img.paste(copy,(col*crop.width, row*crop.height))

img.show()
img.save("manyMountains.png")
```

### Resize, flip, rotate

```python
from PIL import Image
img = Image.open("landscape.jpg")
klein = img.resize((int(img.width/2), int(img.height/2)))
klein.save("klein.jpg")

# klein.rotate(22).save("kleinRotate22.jpg")
klein.transpose(Image.FLIP_LEFT_RIGHT)
klein
```

### Werken met kleuren

```python
from PIL import Image

img = Image.open("phaedra.jpg")

for x in range(0, img.width):
    for y in range(0, img.height):
        r,g,b = img.getpixel((x,y))
        img.putpixel((x,y), (255-r,255-g,255-b))

img.save("phaedraInvert.jpg")
img = Image.open("phaedra.jpg")

for x in range(0, img.width):
    for y in range(0, img.height):
        r,g,b = img.getpixel((x,y))
        img.putpixel((x,y), (r,255-g,0))

img
```

```python
from PIL import Image, ImageChops, ImageOps

img = Image.open("klein.jpg")
ImageChops.invert(img).save("kleinInvert.jpg")
ImageOps.grayscale(img).save("kleinGrayscale.jpg")
```
